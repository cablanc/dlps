```python
import torch
import numpy as np
import matplotlib.pylab as plt

from utils import apply_linear_dynamics, compute_trajectory, LQR_policy, apply_linear_dynamics
```

# System Identification

## Define system variables A and B; and initial conditions x0

```python
# http://www.argmin.net/2018/02/08/lqr/
m = 1.
A_gt = np.array([[1., 1.],
                 [0., 1.]])
B_gt = np.array([[0],
                 [1./m]])

# arbitrarily chosen initial conditions
x0 = np.array([[1.],
              [0.]])

np.random.seed(10)
```

## Generate training data using random ctrl input u_t

```python
dataset_size = 100
# generate control input [0, 1]
u_t = np.random.rand(dataset_size)
# normalize to [-1, 1]
u_t -= 0.5
u_t /= 0.5
# reshape
u_t = u_t.reshape(-1, 1, 1)
# plt.hist(u_t)
```

```python
x_t = [x0]

for t in range(dataset_size):
    x_cur = x_t[-1]
    u_cur = u_t[t]
    x_next = apply_linear_dynamics(A_gt, B_gt, x_cur, u_cur)
    x_t.append(x_next)
    
x_cur = x_t[:-1]
x_next = x_t[1:]
u = u_t
```

```python
x_coord1 = [x_t[t][0] for t in range(dataset_size + 1)]
x_coord2 = [x_t[t][1] for t in range(dataset_size + 1)]

plt.plot(x_coord1)
plt.plot(x_coord2)
```

## Define variables and optimizer

```python
torch.manual_seed(0)

A_hat = torch.rand(2, 2, requires_grad=True)
B_hat = torch.rand(2, 1, requires_grad=True)

num_epochs = 100
batch_size = 10
num_iter = int(num_epochs/batch_size)

lr = 0.1
optim = torch.optim.Adam([A_hat, B_hat], lr=lr)
```

## Estimate linear dynamics by Adam descent

```python
for epoch in range(num_iter):
    idx = np.random.permutation(dataset_size)[:batch_size]
    
    x_cur_batch = torch.Tensor(x_cur)[idx]
    x_next_batch = torch.Tensor(x_next)[idx]
    u_batch = torch.Tensor(u)[idx]
    # print(x0_batch.shape, x1_batch.shape, u_batch.shape)
    
    x_next_pred = torch.matmul(A_hat, x_cur_batch) + torch.matmul(B_hat, u_batch)
    # print(x1_pred.shape)
    
    loss = ((x_next_pred - x_next_batch)**2).sum()
    
    optim.zero_grad()
    loss.backward()
    optim.step()
```

```python
A_hat_np = A_hat.data.numpy()
B_hat_np = B_hat.data.numpy()
```

## Define LQR cost

```python
# cost = \sum_t (x_t - x_goal)**2 + u_t**2
Q = np.eye(2)
Q_f = np.eye(2) * 1.
R = np.array([[10.]])

R_small = np.array([[0.3]])
Q_f_large = np.eye(2) * 100.
```

```python
T = 5
policy = LQR_policy(A_hat_np, B_hat_np, R, Q, Q_f, T)
policy_small_R = LQR_policy(A_hat_np, B_hat_np, R_small, Q, Q_f, T)
policy_large_Q_f = LQR_policy(A_hat_np, B_hat_np, R, Q, Q_f_large, T)

x, u = compute_trajectory(A_gt, B_gt, policy, apply_linear_dynamics, x0, T)
x_small_R, u_small_R = compute_trajectory(A_gt, B_gt, policy_small_R, apply_linear_dynamics, x0, T)
x_large_Q_f, u_large_Q_f = compute_trajectory(A_gt, B_gt, policy_large_Q_f, apply_linear_dynamics, x0, T)
```

```python
x1 = [x_t[0] for x_t in x]
x2 = [x_t[1] for x_t in x]

x1_small_R = [x_t[0] for x_t in x_small_R]
x2_small_R = [x_t[1] for x_t in x_small_R]

x1_large_Q_f = [x_t[0] for x_t in x_large_Q_f]
x2_large_Q_f = [x_t[1] for x_t in x_large_Q_f]

plt.figure(figsize=[10,6])
plt.plot(x1, '-', color='b')
plt.plot(x2, '-', color='tab:orange')

plt.plot(x1_small_R, ':', color='b')
plt.plot(x2_small_R, ':', color='tab:orange')

plt.plot(x1_large_Q_f, '--', color='b')
plt.plot(x2_large_Q_f, '--', color='tab:orange')
plt.legend(['x1', 'x2', 'x1_small_R', 'x2_small_R', 'x1_large_Q_f', 'x2_large_Q_f'])

plt.plot([0]*T, '-', color='lightgrey')
```

```python
u1 = [u_t[0] for u_t in u]
u1_small_R = [u_t[0] for u_t in u_small_R]
u1_large_Q_f = [u_t[0] for u_t in u_large_Q_f]

plt.figure(figsize=[10,6])
plt.plot(u1, '-', color='m')
plt.plot(u1_small_R, ':', color='m')
plt.plot(u1_large_Q_f, '--', color='m')
plt.legend(['u', 'u_small_R', 'u_large_Q_f'])

plt.plot([0]*(T-1), '-', color='lightgrey')
```
